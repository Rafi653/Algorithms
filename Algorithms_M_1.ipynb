```python

import matplotlib.pyplot as plt 
import numpy as np 
# f(n)=n^1.01 ; g(n) = n*(logn)^2

l1 = [(pow(n, 1.01)) for n in range(1000)]
l2 = [n*(pow(np.log(n), 2)) for n in range(1000)]

plt.plot(l1, color ='tab:blue', label='f(n)')
plt.plot(l2, color ='tab:red', label='g(n)') 
plt.legend()
plt.show()

# Big O
```

```python

import matplotlib.pyplot as plt 
import numpy as np 
# f(n)= n^2/log(n) ; g(n)=n(log(n))^2

l1 = [(n*n)/np.log(n) for n in range(1000)]
l2 = [n*(pow(np.log(n), 2)) for n in range(1000)]

plt.plot(l1, color ='tab:blue', label='f(n)')
plt.plot(l2, color ='tab:red', label='g(n)') 
plt.legend()
plt.show()


# Big Omega
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
# f(n)= (log(n))^(log(n)); g(n)=(2)^((log(n))^2)

l1 = [pow(np.log(n), np.log(n)) for n in range(1000)]
l2 = [(pow(2, pow(np.log(n), 2))) for n in range(1000)]

plt.plot(l1, color ='tab:blue', label='f(n)')
plt.plot(l2, color ='tab:red', label='g(n)') 
plt.legend()
plt.show()


# Big O
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

def sigma(first, last, k):
    sum = 0
    for i in range(first, last + 1):
        sum += pow(i, k) 
    return sum

for k in [-2,0,2]:
  l1 = [sigma(1, n, k) for n in range(1, 1000)]
  l2 = [pow(n, k+1) for n in range(1, 1000)]

  plt.plot(l1, color ='tab:blue', label='f(n)')
  plt.plot(l2, color ='tab:red', label='g(n)') 
  plt.legend()
  plt.title(f"for K value: {k}")
  plt.show()

# Depends on constant value, 
```
